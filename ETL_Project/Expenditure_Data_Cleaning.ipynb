```python
#import dependencies
import pandas as pd
```

```python
#read in csv
path = 'Resources/Expenditures_Year_AllAreas_1997_2021.csv'
spending_df = pd.read_csv(path)
```

```python
#Drop extraneous columns
spending_df_reducedcols = spending_df[['GeoName', 'Description', '2019', '2020', '2021']]
spending_df_reducedcols
```

```python
#Use loc function to drop rows containing national-level data and regional aggregations

spending_df_clean = spending_df_reducedcols.loc[(spending_df_reducedcols['GeoName']!='United States') &
                                                (spending_df_reducedcols['GeoName']!='New England') &
                                                (spending_df_reducedcols['GeoName']!= 'Mideast') &
                                                (spending_df_reducedcols['GeoName']!= 'Great Lakes') &
                                                 (spending_df_reducedcols['GeoName']!= 'Plains') &
                                                  (spending_df_reducedcols['GeoName']!= 'Southeast') &
                                                   (spending_df_reducedcols['GeoName']!= 'Southwest') & 
                                                (spending_df_reducedcols['GeoName']!= 'Rocky Mountain') &
                                                 (spending_df_reducedcols['GeoName']!= 'Far West') &
                                               (spending_df_reducedcols['GeoName']!= 'NaN')]

#drop rows full of null values

spending_df_clean_nonull = spending_df_clean.dropna(0, how='all')

spending_df_clean_nonull.rename(columns = {'GeoName': 'us_state', 'Description': 'spending_category', '2019': '2019_spending', '2020': '2020_spending', '2021': '2021_spending'}, inplace=True)

spending_df_clean_nonull


```

```python




spending_df_clean_nonull['us_state'] = spending_df_clean_nonull['us_state'].astype(str)

spending_df_clean_nonull




```

```python
spending_df_clean_nonull.to_csv('state_spending_cleaned.csv', index=False)
```
