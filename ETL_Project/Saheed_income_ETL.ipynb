```python
import pandas as pd
import re
```

```python
income_df = pd.read_csv("Resources/Income_AllAreas_1998_2021.csv")
```

```python
income_df
```

```python
income_df_reduced = income_df[['GeoName', 'Description', '2019', '2020', '2021']]
```

```python
income_df_reduced.head()
```

```python
income_df_reduced = income_df_reduced.loc[income_df_reduced['GeoName']!='United States']
income_df_reduced.head()
```

```python
personal_income_df = income_df_reduced.loc[income_df_reduced['Description'] == ' Personal income ']
```

```python
personal_income_df
```

```python
personal_income_df.info()
```

```python
personal_income_df[['2019', '2020', '2021']] = personal_income_df[['2019', '2020', '2021']].astype('float')
```

```python
personal_income_df.info()
```

```python
personal_income_df = personal_income_df.rename(columns={'GeoName': 'us_state', '2019': '2019_income', '2020':'2020_income', '2021':'2021_income'})
```

```python
personal_income_df.drop('Description', axis = 1, inplace = True)
```

```python
personal_income_df
```

```python
personal_income_df.to_csv("Resources/income_all_final_CLEANED.csv", index = False)
```
