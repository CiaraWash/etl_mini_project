```python
import pandas as pd 
```

```python
pd.options.display.max_colwidth = 200


jobs_industrydf = pd.read_csv("Resources/Jobs_Industry_AllAreas_1998_2021.csv")
jobs_industrydf
```

```python
jobs_industrydf = jobs_industrydf.dropna()
```

```python
jobs_industrydf
```

```python
reducedjobs_df = jobs_industrydf[["GeoFIPS", "GeoName","Description","2019", "2020","2021"]]

reducedjobs_df
```

```python
geonames = reducedjobs_df.groupby("GeoName")

geolist = geonames["GeoName"].count()

geolist
```

```python
reducedjobs2_df=reducedjobs_df.loc[(reducedjobs_df['GeoName'] != "Far West") & (reducedjobs_df['GeoName'] != "Great Lakes") & (reducedjobs_df['GeoName'] != "Mideast") & (reducedjobs_df['GeoName'] != "Plains") & (reducedjobs_df['GeoName'] != "Rocky Mountain") & (reducedjobs_df['GeoName'] != "Southeast") & (reducedjobs_df['GeoName'] != "Southwest") & (reducedjobs_df['GeoName'] != "New England") & (reducedjobs_df['GeoName'] != "United States")] 

reducedjobs2_df
```

```python
reducedjobs2_df = reducedjobs2_df.rename(columns={"GeoName": "us_states", "Description": "industry", "2019": "2019_jobs", 
                                                  "2020": "2020_jobs", "2021":"2021_jobs" })  

reducedjobs2_df
```

```python
del reducedjobs2_df['GeoFIPS']

reducedjobs2_df
```

```python
reducedjobs2_df.dtypes
```

```python
#reducedjobs2_df["2019"] = reducedjobs2_df["2019"].astype("int")

reducedjobs2_df['2019_jobs'] = pd.to_numeric(reducedjobs2_df['2019_jobs'], errors='coerce')
reducedjobs2_df['2020_jobs'] = pd.to_numeric(reducedjobs2_df['2020_jobs'], errors='coerce')
reducedjobs2_df['2021_jobs'] = pd.to_numeric(reducedjobs2_df['2021_jobs'], errors='coerce')

reducedjobs2_df.info()
```

```python
reducedjobs2_df.to_csv("Resources/Job_industry_CLEANED.csv")
```
